```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(4, 3), dpi = 100)
plt.xlabel('w')
plt.ylabel('J')
X = np.linspace(-5, 5, 128)
Y = X ** 2
plt.plot(X, Y)
plt.scatter(X[-10], Y[-10], color = 'red')
```

```python
plt.figure(figsize = (4,3), dpi = 100)
plt.xlabel('w')
plt.ylabel('J')
X = np.linspace(-5, 5, 128)
Y = X ** 2
plt.plot(X, Y)
plt.scatter(X[-10], Y[-10], color = 'grey')
plt.scatter(X[-25], Y[-25], color='red')
```
